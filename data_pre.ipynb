```python
# import data
data1 = pd.read_csv("comp.csv")
data1
```

```python
data2 = pd.read_csv("crsp.csv")
data2
```

```python
x1 = data2
x1['DATE'] = pd.to_datetime(x1['date'], format='%Y%m%d')
x1
```

```python
x1["date_merge"] = x1["DATE"].dt.to_period("M")
x1
```

```python
x2 = data1
x2['DATADATE'] = pd.to_datetime(x2['datadate'], format='%Y%m%d')
x2
```

```python
indicator = x1[['TICKER', 'date_merge']]
indicator
```

```python
x2["date_merge"] = x2["DATADATE"].dt.to_period("M")
x2
```

```python
x2_new = indicator.merge(right = x2, left_on = ['TICKER', 'date_merge'], right_on = ['tic', 'date_merge'], how = 'left')
x2_new
```

```python
x2_new = x2_new.ffill()
x2_new
```

```python
def offset_4month(date):
    return(date + relativedelta(months = 4))
```

```python
x2_new.dropna(inplace = True)
```

```python
x2_new['DATADATE']=x2_new['date_merge'].dt.to_timestamp()
```

```python
x2_new
```

```python
x2_new["DATADATE"] = x2_new["DATADATE"].apply(offset_4month)
x2_new
```

```python
x2_new['date_merge'] = x2_new['DATADATE'].dt.to_period("M")
x2_new
```

```python
merged = x1.merge(right = x2_new, left_on = ['TICKER', 'date_merge'], right_on = ['TICKER', 'date_merge'], how = 'inner')
merged
```

```python
merged.groupby('TICKER').ffill().dropna().columns
```

```python
final_data = merged.groupby('TICKER').ffill().dropna()
```

```python
final_data
```

```python
final_data.to_csv("data.csv")
```
