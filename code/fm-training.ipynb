```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("in_sample.csv")
data
```

```python
data.columns
```

```python
data["saleq"]
```

```python
data = data[pd.to_numeric(data.RET, errors='coerce').notnull()]
```

```python
data['RET'] = data['RET'].astype(np.float)
```

```python
data['VOL']
```

```python
# take value we need
data["size"] = data['PRC'] * data['SHROUT']
data['s2p'] = data['saleq'] / data['size']
data['risk_premium'] = data['RET'] - data['mktrf']
```

```python
data_use = data[['tic','PERMNO', 'VOL', 'RET','risk_premium', 'SPREAD', 'size', 's2p', 'epspxq', 'ceqq', 'mktrf', 'smb', 'hml', 'rf',
       'umd','sprtrn', 'date_merge']]
data_use
```

```python
# uniform size
data_use['sizeln'] = np.log(data_use['size'])
```

```python
data_use['sizeln']
```

```python
data_use.replace([np.inf, -np.inf], np.nan, inplace=True)
data_use.dropna(inplace = True)
```

```python
data_use
```

```python
data_use = data_use[data_use['sizeln'] > 0]
```

```python
timelist = np.unique(data_use[['date_merge']])
```

```python
timelist
```

```python
data_by_time = []
for i in timelist:
    datai = data_use[data_use['date_merge'] == i]
    data_by_time.append(datai)
```

```python
factors = ['VOL', 'SPREAD', 'sizeln', 's2p', 'epspxq']
param_est = {'vol':[], 'spread': [], 'sizeln': [], 's2p': [], 'epspxq': []}
for dataframe in data_by_time:
    X = dataframe[['VOL', 'SPREAD', 'sizeln', 's2p', 'epspxq']]
    y = dataframe['risk_premium']
    model = LinearRegression()
    results = model.fit(X, y)
    param_est['vol'].append(results.coef_[0])
    param_est['spread'].append(results.coef_[1])
    param_est['sizeln'].append(results.coef_[2])
    param_est['s2p'].append(results.coef_[3])
    param_est['epspxq'].append(results.coef_[4])
```

```python
param_est = pd.DataFrame(param_est)
```

```python
param_est.mean()
```

```python
param_est.std()
```

```python
param_tstat = param_est.mean() / (param_est.std() * np.sqrt(252))
```
